# Multiple charts
We can see data more, for exmaple thx to col or rows we can see with more details with more variables

```python
nueva = pd.read_csv(r"cancer.csv")
nueva.head()
```

```python
print(nueva.dtypes)
```

```python
sns.relplot(x = 'Area (mean)', y = 'Texture (mean)', data = nueva, hue ='Perimeter (mean)', col = 'Diagnosis', col_wrap=2);
```

# Scatter plot
    We can use with categorical data

    jitter= False, let us to see as barcharts

```python
sns.catplot(x= "Diagnosis", y ="Area (mean)",data = nueva, jitter =False);
```

```python
sns.catplot(x="Diagnosis", y="Area (mean)", data=nueva.query("`Texture (mean)` > 15"), jitter=False)
#TODO: .query let us querying 
```

```python
sns.catplot(x="Diagnosis", y="Area (mean)", data=nueva.query("`Texture (mean)` > 15"), kind="swarm")
```

# BOX PLOT
Catergorical, distribution with outliers

```python
sns.catplot(x="Diagnosis", y="Area (mean)", data= nueva, kind='box')
```
